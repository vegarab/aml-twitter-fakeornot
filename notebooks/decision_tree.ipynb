```
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.metrics import confusion_matrix, classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

from tfn.preprocess import Dataset
from tfn.feature_extraction.tf_idf import get_tfidf_model
```

```
data = Dataset('twitter')

def _tokenizer(text): return text

vectorizer = CountVectorizer(tokenizer=_tokenizer, lowercase=False)
X_train = vectorizer.fit_transform(data.X_train)
X_test = vectorizer.transform(data.X_test)
```

```
dt = DecisionTreeClassifier(max_depth=None)
dt.fit(X_train, data.y_train)
y_pred = dt.predict(X_test)

print(classification_report(data.y_test, y_pred))
print(confusion_matrix(data.y_test, y_pred))
```

```
from IPython.display import SVG, display
from graphviz import Source

graph = Source(
    tree.export_graphviz(
        dt,
        out_file=None,
        feature_names=vectorizer.get_feature_names(),
        class_names=['Non-disaster', 'Disaster'],
        filled = True)
    )
display(SVG(graph.pipe(format='svg')))
```
